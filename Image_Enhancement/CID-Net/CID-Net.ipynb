```python
# train_E

import shutil
import os

# 디렉토리 경로
output_dir = "/content/output/custome"

# 디렉토리 삭제 (존재하면 삭제)
if os.path.exists(output_dir):
    shutil.rmtree(output_dir)

# 디렉토리 다시 생성
os.makedirs(output_dir)

print(f"Directory {output_dir} has been reset.")

# 테스트
!python /content/drive/MyDrive/HVI-CIDNet/eval.py --unpaired --custome --custome_path /content/drive/MyDrive/Low_Brightness_Extreme/train_E/ --unpaired_weights /content/drive/MyDrive/HVI-CIDNet/weights/w_perc.pth --alpha 0.9

# 드라이브에 저장할 폴더 경로
drive_output_dir = "/content/drive/MyDrive/Low_Brightness_Extreme/train_E_CIDNet"

# /content/output/custome 아래의 모든 .jpg 파일 복사
source_dir = "/content/output/custome"
for file_name in os.listdir(source_dir):
    if file_name.endswith(".jpg"):  # .jpg 파일만 복사
        shutil.copy(os.path.join(source_dir, file_name), drive_output_dir)

print(f"All .jpg files have been copied to {drive_output_dir}")
```
