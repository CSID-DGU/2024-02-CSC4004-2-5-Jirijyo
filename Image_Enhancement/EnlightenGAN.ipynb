```python
# train M

import os
import cv2
from enlighten_inference import EnlightenOnnxModel

# 입력 폴더와 출력 폴더 경로 설정
input_folder = r'C:\Users\yeoch\CSC4004\cctv-people-1\train\Low_Brightness_Moderate\images'
output_folder = r'C:\Users\yeoch\CSC4004\cctv-people-1\train\GAN_Moderate'

# 출력 폴더가 존재하지 않으면 생성
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 모델 초기화
model = EnlightenOnnxModel()

# 입력 폴더의 모든 이미지 처리
for filename in os.listdir(input_folder):
    input_path = os.path.join(input_folder, filename)
    
    # 파일이 이미지인지 확인
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
        # 이미지 읽기
        img = cv2.imread(input_path)
        
        if img is None:
            print(f"Failed to read image: {input_path}")
            continue
        
        # 모델을 사용하여 이미지 처리
        processed = model.predict(img)
        
        # 처리된 이미지를 출력 폴더에 저장
        output_path = os.path.join(output_folder, filename)
        cv2.imwrite(output_path, processed)
        print(f"Processed and saved: {output_path}")

print("모든 이미지 처리가 완료되었습니다.")
```
